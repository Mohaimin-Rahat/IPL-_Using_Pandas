```python
def top_run(batsman_name):
    bat=delivery[delivery["batsman"]==batsman_name]
    return bat.groupby('bowling_team')["batsman_runs"].sum().sort_values(ascending=False)
```

```python
top_run("V Kohli")
```

```python
top_run("DA Warner")
```

```python
death_over=delivery[delivery["over"]>15]
```

```python
death_over.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False).head(10)
```

```python
all_bat=death_over.groupby("batsman")["batsman_runs"].count()
```

```python
all_bat.shape[0]
```

```python
x=all_bat>200
```

```python
all_bat[x].shape[0]
```

```python
batsman_list=all_bat[x].index.tolist()
```

```python
batsman_list
```

```python
bat_43=death_over[death_over["batsman"].isin(batsman_list)]
```

```python
bat_43.shape[0]
```

```python
runs=bat_43.groupby('batsman')["batsman_runs"].sum()
```

```python
balls=bat_43.groupby('batsman')["batsman_runs"].count()
```

```python
sr=runs/balls*100
```

```python
sr.sort_values(ascending=False).head(10)
```
